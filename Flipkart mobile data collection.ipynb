```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from time import sleep
from random import randint
```

```python
headers = {"Accept-Language":"en-US,en;q=0.5"}
```

# Collecting Pages

Flipkart have 42 pages having information about Mobile.
so using numpy we can create array which store value upto 42, starting from 1.

```python
pages = np.arange(1,42,1)
```

# Collecting Data

Making Empty list to store Data.
Using for loop, with Beautiful Soup, Extracting mobile data from Flipkart Website.

```python
price = []
description = []
ram = []
storage = []
display = []
camera = []
battery = []
rating = []
```

```python
for page in pages:
    url = "https://www.flipkart.com/search?q=mobiles&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_1_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_1_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=mobiles%7CMobiles&requestId=5e1eadbc-bfed-4d72-acf9-dc7e57b6d409&page="+str(page)
    result = requests.get(url,headers = headers)
    soup = BeautifulSoup(result.text, "html.parser")
    mob_div = soup.find_all("div", class_="_1AtVbE col-12-12")
    sleep(randint(2,10))
    for mobsection in mob_div:
        desc = soup.find_all("div", class_="_4rR01T")
        for i in range(len(desc)):
            description.append(desc[i].text)
        info = soup.find_all("li", class_="rgWa7D")
        for i in range(0,len(info)):
            p = info[i].text
            if ("RAM" in p):
                ram.append(p)
            elif ("ROM" in p):
                storage.append(p)
            elif ("Display" in p):
                display.append(p)
            elif("Camera" in p):
                camera.append(p)
            elif("Battery" in p):
                battery.append(p)
        mob_prices = soup.find_all("div", class_="_30jeq3 _1_WHN1")
        for i in range(len(mob_prices)):
            price.append(mob_prices[i].text)
        
        rate = soup.find_all("div", class_="_3LWZlK")
        for i in range(len(rate)):
            rating.append(rate[i].text)
    
```

# Converting Data into DataFrame

printing length of each stored list.
we can observe that list doesn't contain equal amount of data.
'rating' having highest length while 'storage' having lowest length.
so we have to slice data before converting it into dataframe.

```python
print(len(price))
print(len(description))
print(len(ram))
print(len(storage))
print(len(display))
print(len(camera))
print(len(battery))
print(len(rating))
```

```python
mob_data = {'Description': description[slice(1340)],
            'Ram':ram[slice(1340)],
            'Storage':storage[slice(1340)],
            'Display':display[slice(1340)],
            'Camera':camera[slice(1340)],
            'Battery':battery[slice(1340)],
            'Rating':rating[slice(1340)],
            'Price': price[slice(1340)]
           }
```

```python
dataset = pd.DataFrame(data = mob_data)
```
