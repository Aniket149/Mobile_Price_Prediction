# The price have strong correlation with ram, storage and front camera

```python
data.head()
```

```python
data['Storage'].unique()
```

```python
sns.set(rc = {'figure.figsize':(9,6)})
sns.scatterplot(data['RAM'],
                data['Price'],
                hue = data['Storage'],
                palette = ['black','darkblue','purple'])
plt.title("Relationship Between Mobile Storage system and Price of the Mobile", fontsize = 14, weight = "bold")
plt.xlabel("RAM", weight = "bold")
plt.ylabel("Price",weight = "bold")
plt.savefig('mobile_storage_price_relationship.jpg', bbox_inches='tight',dpi=150)
plt.show()
```

```python
sns.set(rc = {'figure.figsize':(9,6)})
sns.scatterplot(data['Front_Camera'],
                data['Price'],
                hue = data['Storage'],
                palette = ['black','darkblue','purple'])
plt.title("Relationship Between Mobile Front Camera with Storage system and Price of the Mobile", fontsize = 14, weight = "bold")
plt.xlabel("Front Camera", weight = "bold")
plt.ylabel("Price",weight = "bold")
plt.savefig('mobile_front_camera_price_relationship.jpg',bbox_inches='tight',dpi=150)
plt.show()
```

```python
fig =plt.figure(figsize=(9,6))
plt.subplot(1,2,1)
sns.scatterplot(data['RAM'],
                data['Price'])
plt.subplot(1,2,2)
sns.scatterplot(data['Storage'],
                data['Price'])
fig.tight_layout()
```

```python
fig =plt.figure(figsize=(9,6))
plt.subplot(1,2,1)
sns.scatterplot(data['Front_Camera'],
                data['Price'])
plt.subplot(1,2,2)
sns.scatterplot(data['Camera_Types'],
                data['Price'])
fig.tight_layout()
```

# EDA Conclusion

By exploratory data analysis, it is concluded that Price of mobile have strong correlation with RAM, Storage and Front Camera.
The relationship of these three with price of mobile is stronger.
we will select ram, storage and front camera to predict mobile price(Mobile price prediction model)
